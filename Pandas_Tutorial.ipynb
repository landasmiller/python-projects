```python
bios['height_category'] = bios['height_cm'].apply(lambda x: 'short' if x < 170 else 'tall')
bios.head()
#lambda fuction can be helpful when being specific and categorizing the data with string values


def categorize_athlete(row):
    if row['height_cm'] < 175 and row['weight_kg'] < 70:
      return 'Lightweight'
    elif row['height_cm'] < 185 or row['weight_kg'] <= 80:
      return 'Middleweight'

    else:
      return 'Heavyweight'

bios['Category'] = bios.apply(categorize_athlete, axis = 1)

```
